```python
import matplotlib.pyplot as plt
import numpy as np

# Data untuk percobaan
waktu = np.array([0.39, 0.40, 0.41, 0.42])  # Waktu (T) dalam detik
distance = 0.17  # Panjang tali (L) dalam meter
k_gravitasi = 9.85  # Konstanta gravitasi yang digunakan dalam eksperimen

# Perhitungan percepatan gravitasi (g)
g = (k_gravitasi * distance) / (waktu ** 2)

# Membuat grafik
plt.figure(figsize=(8, 5))
plt.plot(waktu, g, marker='o', color='blue', label='Percepatan Gravitasi (g)')
plt.axhline(y=9.81, color='red', linestyle='--', label='Nilai Ideal (9.81 m/s²)')
plt.title('Percepatan Gravitasi Berdasarkan Waktu Osilasi', fontsize=14)
plt.xlabel('Waktu (T) dalam detik', fontsize=12)
plt.ylabel('Percepatan Gravitasi (g) dalam m/s²', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.show()
```

```python
import pandas as pd
import math

# Data waktu osilasi (T) dan panjang tali (L)
waktu_osilasi = [0.39, 0.40, 0.41, 0.42]  # dalam detik
panjang_tali = [0.17, 0.17, 0.17, 0.17]  # dalam meter

# Fungsi untuk menghitung percepatan gravitasi g
def hitung_gravitasi(T, L):
    return (4 * math.pi**2 * L) / (T**2)

# Menghitung percepatan gravitasi untuk setiap waktu osilasi
percepatan_gravitasi = [hitung_gravitasi(T, L) for T, L in zip(waktu_osilasi, panjang_tali)]

# Membuat DataFrame dengan pandas
data = {
    'Waktu Osilasi (T) [s]': waktu_osilasi,
    'Panjang Tali (L) [m]': panjang_tali,
    'Percepatan Gravitasi (g) [m/s²]': percepatan_gravitasi
}

df = pd.DataFrame(data)

# Menampilkan tabel
print(df)
```
